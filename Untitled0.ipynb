```python
# STEP 1: Upload file to Colab
from google.colab import files
print("Please upload: health_ai_whatsapp_100_conversations_long.txt")
uploaded = files.upload()

# Get the actual filename
filename = list(uploaded.keys())[0]
print(f"✅ Uploaded: {filename}\n")

# STEP 2: Now run the analysis
import re
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 10)

print("="*80)
print("HEALTHCARE CONVERSATION RISK ANALYSIS - Prince Nyuswa")
print("="*80 + "\n")

# Load conversations
def load_conversations(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()

    conversations = []
    blocks = content.split("========== Conversation ==========")

    for block in blocks:
        if not block.strip():
            continue
        exchanges = []
        for line in block.strip().split('\n'):
            match = re.match(r'\[([^\]]+)\]\s+(User|AI):\s+(.+)', line)
            if match:
                _, speaker, message = match.groups()
                exchanges.append({'speaker': speaker, 'message': message.strip()})

        if exchanges:
            conversations.append({
                'id': len(conversations) + 1,
                'exchanges': exchanges,
                'user_messages': [e['message'] for e in exchanges if e['speaker'] == 'User']
            })
    return conversations

conversations = load_conversations(filename)  # Use the uploaded filename
print(f"✅ Loaded {len(conversations)} conversations\n")

# HIV Risk Scorer
class HIVRiskScorer:
    def __init__(self):
        self.keywords = {
            'fever': 15, 'sweat': 10, 'rash': 12, 'tired': 6, 'weight loss': 15,
            'partner': 12, 'unprotected': 30, 'exposure': 25, 'condom': 15,
            'hiv': 20, 'positive': 25, 'test': 8, 'urgent': 15, 'worried': 8
        }

    def score(self, conversation):
        text = ' '.join(conversation['user_messages']).lower()
        score = min(100, sum(w for k, w in self.keywords.items() if k in text))

        if score >= 76: level, urgency = "CRITICAL", "Immediate"
        elif score >= 51: level, urgency = "HIGH", "Within 3 days"
        elif score >= 26: level, urgency = "MODERATE", "Within 2 weeks"
        else: level, urgency = "LOW", "Routine"

        return {'score': score, 'level': level, 'urgency': urgency}

# Mental Health Risk Scorer
class MentalHealthRiskScorer:
    def __init__(self):
        self.keywords = {
            'sad': 10, 'depressed': 18, 'hopeless': 20, 'worthless': 20,
            'worried': 10, 'anxious': 15, 'panic': 18, 'stressed': 12,
            'suicide': 35, 'kill myself': 40, 'die': 25, 'death': 18
        }
        self.crisis_words = ['suicide', 'kill myself', 'die', 'death']

    def score(self, conversation):
        text = ' '.join(conversation['user_messages']).lower()
        score = min(100, max(0, sum(w for k, w in self.keywords.items() if k in text)))
        crisis = any(w in text for w in self.crisis_words)

        if crisis or score >= 76: level, urgency = "CRITICAL", "Immediate crisis"
        elif score >= 51: level, urgency = "HIGH", "Within 1 week"
        elif score >= 26: level, urgency = "MODERATE", "Counseling needed"
        else: level, urgency = "LOW", "Wellness"

        return {'score': score, 'level': level, 'urgency': urgency, 'crisis': crisis}

# Treatment Recommender
class TreatmentRecommender:
    def recommend(self, hiv_risk, mh_risk):
        recs = {'hiv': [], 'mental_health': [], 'resources': []}

        if hiv_risk['level'] == "CRITICAL":
            recs['hiv'] = ["🚨 URGENT: HIV rapid test within 24 hours", "Consider PEP if exposure within 72 hours"]
        elif hiv_risk['level'] == "HIGH":
            recs['hiv'] = ["HIV testing within 3 days", "Visit nearest clinic (free testing)"]
        elif hiv_risk['level'] == "MODERATE":
            recs['hiv'] = ["HIV testing within 2 weeks", "Partner testing recommended"]
        else:
            recs['hiv'] = ["Routine annual HIV screening"]

        if mh_risk['crisis'] or mh_risk['level'] == "CRITICAL":
            recs['mental_health'] = ["🚨 CRISIS: Call SADAG: 0800 567 567", "Emergency services: 10111"]
        elif mh_risk['level'] == "HIGH":
            recs['mental_health'] = ["Mental health professional within 1 week", "Visit Community Health Centre"]
        elif mh_risk['level'] == "MODERATE":
            recs['mental_health'] = ["Counseling within 2 weeks"]
        else:
            recs['mental_health'] = ["Wellness resources available"]

        recs['resources'] = ["AIDS Helpline: 0800 012 322", "SADAG: 0800 567 567", "Lifeline: 0861 322 322"]
        return recs

# Analyze all conversations
print("Running risk analysis...\n")
hiv_scorer = HIVRiskScorer()
mh_scorer = MentalHealthRiskScorer()
recommender = TreatmentRecommender()

for conv in conversations:
    conv['hiv_risk'] = hiv_scorer.score(conv)
    conv['mh_risk'] = mh_scorer.score(conv)
    conv['recommendations'] = recommender.recommend(conv['hiv_risk'], conv['mh_risk'])

# Create DataFrame
df = pd.DataFrame([{
    'id': c['id'],
    'hiv_score': c['hiv_risk']['score'],
    'hiv_level': c['hiv_risk']['level'],
    'mh_score': c['mh_risk']['score'],
    'mh_level': c['mh_risk']['level'],
    'crisis': c['mh_risk']['crisis']
} for c in conversations])

print("✅ Analysis complete!\n")
print("STATISTICS:")
print(f"  HIV Risk - Mean: {df['hiv_score'].mean():.1f}, Median: {df['hiv_score'].median():.1f}")
print(f"  Mental Health - Mean: {df['mh_score'].mean():.1f}, Median: {df['mh_score'].median():.1f}")
print(f"  Crisis Cases: {df['crisis'].sum()} ({df['crisis'].sum()/len(df)*100:.1f}%)\n")

# Visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

axes[0,0].hist(df['hiv_score'], bins=20, color='#e74c3c', alpha=0.7, edgecolor='black')
axes[0,0].axvline(df['hiv_score'].mean(), color='darkred', linestyle='--', linewidth=2)
axes[0,0].set_title('HIV Risk Score Distribution', fontsize=13, fontweight='bold')

axes[0,1].hist(df['mh_score'], bins=20, color='#3498db', alpha=0.7, edgecolor='black')
axes[0,1].axvline(df['mh_score'].mean(), color='darkblue', linestyle='--', linewidth=2)
axes[0,1].set_title('Mental Health Risk Distribution', fontsize=13, fontweight='bold')

risk_comp = pd.DataFrame({
    'HIV': df['hiv_level'].value_counts(),
    'Mental Health': df['mh_level'].value_counts()
}).fillna(0)
risk_comp.plot(kind='bar', ax=axes[1,0], color=['#e74c3c', '#3498db'])
axes[1,0].set_title('Risk Level Comparison', fontsize=13, fontweight='bold')
axes[1,0].set_xticklabels(axes[1,0].get_xticklabels(), rotation=45, ha='right')

axes[1,1].scatter(df['hiv_score'], df['mh_score'], c=df['crisis'], cmap='RdYlGn_r', alpha=0.6, s=100)
axes[1,1].set_title('Risk Score Correlation', fontsize=13, fontweight='bold')
correlation = df['hiv_score'].corr(df['mh_score'])
axes[1,1].text(0.05, 0.95, f'Correlation: {correlation:.3f}', transform=axes[1,1].transAxes,
               bbox=dict(boxstyle='round', facecolor='white'))

plt.suptitle('Healthcare Risk Analysis Dashboard - Prince Nyuswa', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('risk_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()
print("✅ Visualization created\n")

# Export results
df.to_csv('risk_analysis_results.csv', index=False)
print("✅ Results exported\n")

# High-risk report
high_risk = df[(df['hiv_level'].isin(['HIGH','CRITICAL'])) | (df['mh_level'].isin(['HIGH','CRITICAL']))]
print(f"🚨 HIGH-RISK CASES: {len(high_risk)}\n")

if len(high_risk) > 0:
    print("SAMPLE HIGH-RISK CASES:")
    for _, row in high_risk.head(5).iterrows():
        conv = conversations[int(row['id'])-1]
        print(f"\nConversation {row['id']}: HIV {row['hiv_level']} ({row['hiv_score']}) | MH {row['mh_level']} ({row['mh_score']})")
        if row['crisis']: print("  🚨 CRISIS FLAG")
        print(f"  Actions: {conv['recommendations']['hiv'][0]} | {conv['recommendations']['mental_health'][0]}")

print("\n" + "="*80)
print("COMPLETE | Prince Nyuswa | 083 200 5622")
print("="*80)

# Download files
from google.colab import files
files.download('risk_dashboard.png')
files.download('risk_analysis_results.csv')
```
